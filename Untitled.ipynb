Beautiful Soup is a library that makes it easy to scrape information from web pages.

```python
from bs4 import BeautifulSoup as BS
```

pandas is a library in python used to open and manipulate dataset. It is used here to create dataframe.

```python
import pandas as pd
```

```python
wiki = "https://www.forbesindia.com/article/explainers/1-usd-to-inr-history/88635/1#:~:text=In%20the%20early%20years%20of,about%20%E2%82%B980%20in%202023."
```

The prettify() method in Beautiful Soup is used to visualize and understand the structure of HTML and XML documents.

```python
page = ul(wiki)
soup = BS(page)
print(soup.prettify)
```

The find_all method is used for finding out all tags with the specified tag name or id and returning them as a list of type bs4

```python
all_tables=soup.find_all('table')
```

The find method is used for finding out the first tag with the specified name or id and returning an object of type bs4.

```python
right_table=soup.find('table',style="border-collapse: collapse;")
right_table
```

```python
data=right_table.find_all("td")
data
```

Adding units in list using append operator.

```python
A=[]
for i in range(0,52,++2): 
    A.append(data[i].find(text=True))       
```

```python
B=[]
for j in range(1,52,++2):
    B.append(data[j].find(text=True)) 
```

Creating data frame using pandas operator

```python
df=pd.DataFrame(A,columns=['Year'])
df["Exchange Rate"]=B
```
